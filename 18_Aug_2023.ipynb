```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
data,labels = make_blobs(n_samples=300,centers=4,cluster_std=1.0,random_state=0)
```

```python
plt.scatter(data[:, 0],data[:, 1],c=labels,cmap='viridis')
plt.title("Original Data")
plt.show()
```

```python
#performe agglomerative clustering
```

```python
a1 = AgglomerativeClustering(n_clusters=3)
```

```python
labels = a1.fit_predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1],c=labels,cmap='viridis')
plt.title("Agglomerative Clustering")
plt.show()
```

```python
linked = linkage(data , 'ward')
dendrogram(linked)
plt.title("cluster Dendrogram")
plt.show()
```
